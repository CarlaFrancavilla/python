## Ejercicio 1 - programa para ingresar temperatura en Celcius y convertirla a Fahrenheit

```python

C = float(input("ingrese aquí la temperatura en Celsius: "))
```

```python
F = C * (9 / 5) + 32 
```

```python
print(f"la temperatura {round (C, 2)} es {round (F, 2)} en grados fahrenheit")
```

## Ejercicio 2 - Escribir un programa que reciba 3 números y calcule el promedio

### Pedir de a un número, sumar números, dividir, imprimir resultado

### Pero hay otra forma...

```python
numeros = input("Ingrese los tres números con lo que se calcula el promedio, separar por comas: ")
```

```python
lista_numeros = numeros.split(",")
total = float(lista_numeros[0]) + float(lista_numeros[1]) + float(lista_numeros[2])
```

```python
promedio = total / len(lista_numeros)
print(f"El promedio de los números ingresados es {round(promedio, 2)}")
```

## Ejercicio 3 - Calculo de salario semanal

### Programa donde se pueda ingresar el monto de las ventas de un vendedor y calcular el salario total, incluyendo las comisiones que serán del 10% sobre sus ventas

```python
salario = int(500)
ventas = int(input("cuántas ventas hiciste esta semana?: "))
```

```python
comision = 0.10 * ventas
total_salario = salario + comision
print(f"El salario de esta semana es {round(total_salario, 2)}")
```
